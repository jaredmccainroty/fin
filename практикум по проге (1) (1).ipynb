```python
pygame.init()
screen = pygame.display.set_mode((800, 600) , pygame.RESIZABLE)
screen.fill([255, 255, 255])
pygame.display.set_caption('Первая программа в pygame')

pygame.draw.circle(screen, 'red', [200, 100], 30, width = 0)
pygame.draw.circle(screen, [255, 154, 13], [100, 400], 50, width = 15)
pygame.draw.circle(screen, '#FFEE54', [400,300], 100, width = 5)

pygame.draw.rect(screen, 'yellow', [400,20,300,200], 0)

for i in range(5):
    top = random.randint(50, 700)
    left = random.randint(50, 500)
    w = random.randint(10, 200)
    h = random.randint(10, 100)
    color = [random.randint(0,255), random.randint(0, 255), random.randint(0,255)]
    pygame.draw.rect(screen, color, [top, left, w, h], 4)


dots=[[221, 432], [225, 331], [133,342], [141,310],
      [51,230], [74,217], [58,153], [114, 164], [123, 135], 
      [176,190], [159,77], [193,93], [230,28], [267,93], [301,77], 
      [284, 190], [327,135], [336,164], [402, 153], [386,217], [409,230], 
      [319, 310], [327, 342], [233,331], [237,432]]
pygame.draw.lines(screen, 'green',True, dots,2) #closed=True первая и последняя точка соединены
#яблоко на экране
apple=pygame.image.load('apple.png')
screen.blit(apple,[400,450]) #копирование пикселей с растрового изображения (блиттинг) pygame.display.flip() #обновляет монитор
#передвинуть изображение в новые координаты pygame.time.delay(2000)# 2000мс=2 сек
pygame.draw.rect(screen, 'white',[400,450,150,150],0) #стираем старое яблоко
screen.blit(apple,[600,450]) #копирование пикселей с растрового изображения (блиттинг)
center_x = 400  # центр экрана по горизонтали
center_y = 300  # центр экрана по вертикали
house_width = 200
house_height = 150
roof_height = 100

# Координаты точек для домика
house_points = [
    (center_x - house_width // 2, center_y + house_height // 2), 
    (center_x - house_width // 2, center_y - house_height // 2),  
    (center_x, center_y - house_height // 2 - roof_height),      
    (center_x + house_width // 2, center_y - house_height // 2),  
    (center_x + house_width // 2, center_y + house_height // 2),  
    (center_x - house_width // 2, center_y + house_height // 2)  
]

# Рисуем домик
pygame.draw.lines(screen, 'brown', True, house_points, 2)
pygame.display.flip() #обновляет монитор
running=True
while running:
    for event in pygame.event.get():
        if event.type==pygame.QUIT: #пришло ли событие нажатия на крестик
            running=False
pygame.quit() #закрыть окно крестиком
```

```python
pygame.init()

# Настройки окна
width, height = 800, 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption('Анимация фигур')

# Цвета
colors = ['red', 'green', 'blue', 'yellow', 'purple', 'orange']

# Начальные параметры фигур
shapes = [
    {'type': 'rect', 'color': 'red', 'x': 100, 'y': 100, 'width': 50, 'height': 50, 'speed': 5},
    {'type': 'circle', 'color': 'green', 'x': 200, 'y': 200, 'radius': 30, 'speed': 5},
    {'type': 'triangle', 'color': 'blue', 'x': 300, 'y': 300, 'points': [(0, 0), (50, 0), (25, 50)], 'speed': 5},
    {'type': 'rect', 'color': 'yellow', 'x': 400, 'y': 400, 'width': 80, 'height': 40, 'speed': 5}
]

def draw_shape(shape):
    if shape['type'] == 'rect':
        pygame.draw.rect(screen, shape['color'], (shape['x'], shape['y'], shape['width'], shape['height']))
    elif shape['type'] == 'circle':
        pygame.draw.circle(screen, shape['color'], (shape['x'], shape['y']), shape['radius'])
    elif shape['type'] == 'triangle':
        points = [(shape['x'] + x, shape['y'] + y) for x, y in shape['points']]
        pygame.draw.polygon(screen, shape['color'], points)

def change_color(shape):
    shape['color'] = random.choice(colors)

def check_boundary(shape):
    if shape['type'] == 'rect':
        if shape['x'] <= 0 or shape['x'] + shape['width'] >= width:
            shape['speed'] *= -1
            change_color(shape)
    elif shape['type'] == 'circle':
        if shape['x'] - shape['radius'] <= 0 or shape['x'] + shape['radius'] >= width:
            shape['speed'] *= -1
            change_color(shape)
    elif shape['type'] == 'triangle':
        min_x = min([shape['x'] + x for x, y in shape['points']])
        max_x = max([shape['x'] + x for x, y in shape['points']])
        if min_x <= 0 or max_x >= width:
            shape['speed'] *= -1
            change_color(shape)

def check_click(shape, pos):
    if shape['type'] == 'rect':
        if shape['x'] <= pos[0] <= shape['x'] + shape['width'] and shape['y'] <= pos[1] <= shape['y'] + shape['height']:
            change_color(shape)
    elif shape['type'] == 'circle':
        if (pos[0] - shape['x']) ** 2 + (pos[1] - shape['y']) ** 2 <= shape['radius'] ** 2:
            change_color(shape)
    elif shape['type'] == 'triangle':
        points = [(shape['x'] + x, shape['y'] + y) for x, y in shape['points']]
        if pygame.Rect(min([x for x, y in points]), min([y for x, y in points]),
                       max([x for x, y in points]) - min([x for x, y in points]),
                       max([y for x, y in points]) - min([y for x, y in points])).collidepoint(pos):
            change_color(shape)

running = True
clock = pygame.time.Clock()

while running:
    screen.fill((255, 255, 255))  # Очистка экрана

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            for shape in shapes:
                check_click(shape, event.pos)

    for shape in shapes:
        shape['x'] += shape['speed']
        check_boundary(shape)
        draw_shape(shape)

    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```
